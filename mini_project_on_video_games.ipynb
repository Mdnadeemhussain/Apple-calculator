# Transformation Function

```python
#Map and Apply
#Change / Modify certain values
#Map is used for small changes (One line code changes)
#Apply is used for larger changes (>1 line code changes)
```

```python
#increse five in Na-sales
video_game.NA_Sales=video_game.NA_Sales.map(lambda x:x+5)
```

```python
video_game.NA_Sales.head(10)
```

```python
#na-sale and eu_sale both have reduce by 5
def sub5(row):
  row.NA_Sales-=5
  row.EU_Sales-=5
  return row

video_game= video_game.apply(sub5, axis=1)
```

```python
video_game.head(10)
```

# Grouping and sorting

```python
# which developer has made hoe much sales in usd globally
video_game.groupby("Developer").Global_Sales.sum().sort_values(ascending=False)
```
